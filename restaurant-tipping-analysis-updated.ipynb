# Restaurant Tipping Analysis

A practical data analysis project exploring tipping behavior in a restaurant setting using Python, Seaborn, and NumPy.

The goal is to practice clear data visualization and extract useful insights about customer tipping behavior.

## Project Goals

- Load and explore the `tips` dataset
- Create different Seaborn visualizations
- Customize plot style, color, and grouping
- Identify simple patterns in tipping behavior

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

## Load Dataset

````python
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
````

## NumPy Analysis

````python
# Calculate tip percentage
tips_dataset['tip_pct'] = tips_dataset['tip'] / tips_dataset['total_bill']

# Basic statistics using numpy
avg_tip = np.mean(tips_dataset['tip'])
max_tip = np.max(tips_dataset['tip'])
min_tip = np.min(tips_dataset['tip'])

print("Average Tip:", avg_tip)
print("Max Tip:", max_tip)
print("Min Tip:", min_tip)
````

**Insight:**  
The average tip gives a general idea of customer behavior, while max and min values show the range of tipping.

````python
# Tip percentage distribution
sns.histplot(tips_dataset['tip_pct'], kde=True)
plt.title('Tip Percentage Distribution')
plt.xlabel('Tip Percentage')
plt.show()
````

**Insight:**  
Most customers tip within a certain percentage range, with fewer extreme values.

## Dataset Overview

````python
print("Dataset shape:", tips_dataset.shape)
tips_dataset.info()
````

````python
tips_dataset.describe()
````

## 1. Histogram: Tip Distribution by Day

This histogram shows the distribution of tip values across different days.

````python
sns.set_style('whitegrid')

sns.histplot(
    data=tips_dataset,
    x='tip',
    hue='day',
    element='step'
)

plt.title('Histogram of Tips by Day')
plt.xlabel('Tip Amount')
plt.ylabel('Count')
plt.show()
````

**Insight:**  
Most tips are relatively small, while only a few customers leave large tips. This means the tip distribution is right-skewed.

## 2. Displot: Day Distribution by Sex

This plot compares the number of male and female customers across different days.

````python
sns.displot(
    data=tips_dataset,
    x='day',
    hue='sex',
    multiple='dodge',
    shrink=0.8,
    palette='Set1',
    hue_order=['Female', 'Male']
)

plt.title('Distribution of Days by Sex')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
````

**Insight:**  
The number of customers differs by day and gender. This helps compare customer distribution across weekdays and weekends.

## 3. ECDF Plot: Tip Distribution by Time

This ECDF plot compares cumulative tip distribution between lunch and dinner.

````python
sns.set_style('darkgrid')
sns.set(font_scale=1.25)

sns.ecdfplot(
    data=tips_dataset,
    x='tip',
    hue='time',
    palette='summer',
    linewidth=3
)

plt.title('ECDF of Tips by Time')
plt.xlabel('Tip Amount')
plt.ylabel('Cumulative Proportion')
plt.show()
````

**Insight:**  
Dinner tends to have higher tip values than lunch. This may be because dinner bills are usually larger or involve more people.

## 4. Scatter Plot: Total Bill vs Tip

This plot shows the relationship between the bill amount and the tip amount.

````python
sns.set_style('whitegrid')
sns.set(font_scale=1)

sns.scatterplot(
    data=tips_dataset,
    x='total_bill',
    y='tip',
    hue='time'
)

plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.show()
````

**Insight:**  
There is a clear positive relationship between total bill and tip. Customers who spend more usually leave higher tips.

## 5. Box Plot: Tips by Day

This plot compares tip amounts across different days.

````python
sns.boxplot(
    data=tips_dataset,
    x='day',
    y='tip'
)

plt.title('Tip Amount by Day')
plt.xlabel('Day')
plt.ylabel('Tip Amount')
plt.show()
````

**Insight:**  
Weekend days show more variation in tip values. This suggests customer spending and tipping behavior may change during busier days.

## Final Analysis

The visualizations show that tipping behavior is mainly affected by:

- **Total bill amount:** higher bills usually lead to higher tips.
- **Meal time:** dinner tends to have higher tip values than lunch.
- **Day of the week:** weekends show more activity and more variation in tips.
- **Distribution shape:** most tips are small, with fewer high-tip values.

Overall, Seaborn makes it easier to compare groups, understand distributions, and identify patterns visually.
